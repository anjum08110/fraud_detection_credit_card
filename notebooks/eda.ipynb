The dataset contains no missing values, indicating that no additional imputation or cleaning is required before modeling.

Fraud transactions account for only approximately 0.17% of all transactions, showing an extreme class imbalance between fraud and non-fraud cases.

This severe class imbalance means that accuracy is not a reliable evaluation metric, and metrics such as precision, recall, F1-score, and PR-AUC are more appropriate for fraud detection.

```python
count = data['Class'].value_counts(normalize=True)
print(count)

count.plot(kind='bar')
plt.title("Class Distribution")
plt.show()
```

```python
data['Amount'].hist(bins=50)
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
data[data['Class']==0]["Amount"].hist(bins=50,alpha=0.7,label='Normal')
data[data['Class']==1]["Amount"].hist(bins=50,alpha=0.7,label='Fraud')
plt.legend()
plt.title('Amount Distribution By Class')
plt.show()
```
